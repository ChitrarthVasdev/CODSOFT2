```python
def chatbot():
    import datetime
    import random
    import pytz

    print("Chatbot: Hello! I'm your smart chatbot. Type 'exit' to end the chat.")

    # Predefined responses
    jokes = [
        "Why don't scientists trust atoms? Because they make up everything!",
        "Why did the computer go to the doctor? Because it had a virus!",
        "Why did the JavaScript developer wear glasses? Because they didn‚Äôt C#!",
        "What do you get if you cross a cat with a dark horse? Kitty Perry.",
        "I told my computer I needed a break, and it said 'No problem ‚Äî I'll go to sleep!'"
    ]

    facts = [
        "Sharks existed before trees!",
        "Oxford University is older than the Aztec Empire.",
        "There are more stars in the universe than grains of sand on Earth.",
        "Octopuses have three hearts!",
        "Honey never spoils ‚Äî edible honey was found in 3,000-year-old Egyptian tombs."
    ]

    quotes = [
        "Believe you can and you're halfway there.",
        "Success is not final, failure is not fatal: It is the courage to continue that counts.",
        "Push yourself, because no one else is going to do it for you.",
        "Your limitation‚Äîit's only your imagination.",
        "Great things never come from comfort zones."
    ]

    faqs = {
        "what is your purpose": "I'm here to assist you with information, answer basic questions, and make your day a little better!",
        "who created you": "I was created by a developer who wanted to explore basic AI and rule-based systems.",
        "can you learn": "I'm a rule-based bot, so I don't learn by myself ‚Äî but I can be updated with new rules!",
        "what language do you speak": "I understand English for now.",
        "do you have emotions": "I can simulate empathy, but I don't truly feel emotions like humans do."
    }

    while True:
        user_input = input("You: ").lower()

        if user_input == "exit":
            print("Chatbot: Goodbye! Take care and come back soon! üëã")
            break

        elif "hello" in user_input or "hi" in user_input:
            print("Chatbot: Hello there! How can I assist you today?")

        elif "how are you" in user_input:
            print("Chatbot: I'm just a bot, but I'm doing great! Thanks for asking üòä")

        elif "your name" in user_input:
            print("Chatbot: I'm ChatBot 101 ‚Äî your friendly assistant.")

        elif "help" in user_input:
            print("Chatbot: You can ask me about the time, date, jokes, quotes, calculations, and more!")

        elif "weather" in user_input:
            print("Chatbot: I can't give live weather updates yet, but I hope it's sunny wherever you are!")

        elif "time" in user_input:
            ist = pytz.timezone('Asia/Kolkata')
            now_ist = datetime.datetime.now(ist)
            print(f"Chatbot: The current IST time is {now_ist.strftime('%H:%M:%S')}.")

        elif "date" in user_input:
            today = datetime.datetime.now(pytz.timezone('Asia/Kolkata')).strftime("%Y-%m-%d")
            print(f"Chatbot: Today's date is {today}.")

        elif "day" in user_input:
            day = datetime.datetime.now(pytz.timezone('Asia/Kolkata')).strftime("%A")
            print(f"Chatbot: Today is {day}.")

        elif "joke" in user_input:
            print("Chatbot:", random.choice(jokes))

        elif "fact" in user_input:
            print("Chatbot:", random.choice(facts))

        elif "quote" in user_input:
            print("Chatbot:", random.choice(quotes))

        elif "thank" in user_input:
            print("Chatbot: You're most welcome! üòä")

        elif "calculate" in user_input:
            try:
                expression = user_input.replace("calculate", "").strip()
                result = eval(expression)
                print(f"Chatbot: The result is {result}")
            except Exception:
                print("Chatbot: Sorry, I couldn't calculate that. Try something like 'calculate 10/2'.")

        elif "good morning" in user_input:
            print("Chatbot: Good morning! Wishing you a bright and productive day! ‚òÄÔ∏è")

        elif "good afternoon" in user_input:
            print("Chatbot: Good afternoon! Hope your day is going well! üåû")

        elif "good evening" in user_input:
            print("Chatbot: Good evening! How can I help you unwind today? üåÜ")

        elif "bye" in user_input or "goodbye" in user_input:
            print("Chatbot: Goodbye! Come chat with me anytime. üëã")

        elif "i am sad" in user_input or "feeling sad" in user_input:
            print("Chatbot: I'm here for you. Remember, tough times don't last, but tough people do. ‚ù§Ô∏è")

        elif "i am happy" in user_input or "feeling happy" in user_input:
            print("Chatbot: That's wonderful to hear! Keep smiling üòÑ")

        else:
            matched = False
            for key_phrase, response in faqs.items():
                if key_phrase in user_input:
                    print("Chatbot:", response)
                    matched = True
                    break

            if not matched:
                print("Chatbot: I'm not sure how to respond to that. Try saying 'help' to see what I can do.")

# Run the chatbot
while True:
    chatbot()
    restart = input("Do you want to chat again? (yes/no): ").strip().lower()
    if restart != 'yes':
        print("Chatbot session ended. üëã")
        break
```
